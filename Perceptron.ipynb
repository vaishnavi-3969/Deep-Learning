```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://github.com/campusx-official/100-days-of-deep-learning/blob/main/day3/placement.csv"
```

```python
df = pd.read_csv("placement.csv")
```

```python
print(df.head)
```

```python
sns.scatterplot(
    df
)
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(x,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.values,y.values,clf=p,legend=2)
```
